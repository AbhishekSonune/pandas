```python
import matplotlib.pyplot as plt
import numpy as np

# Create a purple background
logo = np.ones((100, 300, 3))
logo[:, :] = [0.36, 0.20, 0.60]  # NumPy purple

# Draw a white "N"
logo[20:80, 50:70] = [1, 1, 1]    # Left bar
logo[20:80, 230:250] = [1, 1, 1]  # Right bar
for i in range(20, 81):
    x = 70 + (i-20)*2
    logo[i, x:x+8] = [1, 1, 1]    # Diagonal

plt.figure(figsize=(6, 2))
plt.imshow(logo)
plt.axis('off')
plt.title("NumPy Logo (Generated)")
plt.show()
```
