## Advance operation with Business example


```python
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
# data structure:[resturenat_id,2021,2022,2023,2024]
sales_data =numpy.array([
    [1, 15000, 180000,220000, 250000], #paradise bieyani
    [2, 12000, 140000,160000, 190000], #Beijing bites
    [3, 20000, 230000,260000, 300000], #pizzahub
    [4, 18000, 210000,240000, 270000], #bugger ppoint
    [5, 16000, 185000,205000, 230000] #chai point
])

print("==== ZOmato sales analysis ==== ")
print("\n sales data shape", sales_data.shape)
print("sample data for 1st 3 restue:",sales_data[:3])
```

```python
# total sale par year
print(numpy.sum(sales_data, axis=0))
yeary_total =numpy.sum(sales_data[:, 1:], axis=0) # axis zero is column wise i under stand
print(yeary_total)




#minium sales restarunt 
min_sales = numpy.min(sales_data[:, 1:], axis=1) # the axis is 1 is row wise and  
print("#minium sales restarunt :",min_sales)


# maximum sales per year
max_sales = numpy.max(sales_data[:, 1:], axis=0)
print("maximum sales per year:",max_sales)

#Average sales per resturent 
avg_sales = numpy.mean(sales_data[:, 1:],axis=1)
print("#Average sales per resturent   :",avg_sales)
```

```python
cumsum = numpy.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(numpy.mean(cumsum, axis=0))
plt.title("Average cimsum sales accrosss sale all restaurent")
plt.xlabel("year")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
