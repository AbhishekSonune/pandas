## Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}
df = pd.DataFrame(data)
df["Month"] = df["Date"].dt.month_name()
df["Qurter"] = 'Q' + df["Date"].dt.quarter.astype(str)
df
```

```python
pivot2 = pd.pivot_table(df,values ="Sales",index="Region",columns ="Product") # the automatic used in operating system (mean) count in output this number
pivot2
```

```python
pivot2 = pd.pivot_table(df,values ="Sales",index="Region",columns ="Product",aggfunc ="median") # the calculate (aggfunc) used in function and  sum value
pivot2 
```

```python
pivot2 = pd.pivot_table(df,values =["Sales","Units"],index="Region",columns ="Product") # the multiple value in add (values)
pivot2 
```
